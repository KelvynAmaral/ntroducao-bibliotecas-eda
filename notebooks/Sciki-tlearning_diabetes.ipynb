```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/KelvynAmaral/MLJourney/refs/heads/main/datasets/diabetes.csv'
diabetes = pd.read_csv(url, sep= ',')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = diabetes.drop(['Class variable'], axis=1)
```

```python
x.head()
```

```python
y = diabetes['Class variable']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
len(x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

Passar hiperparâmetros do modelo

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

```python
acuracia = knn.score(x_test, y_test)
```
